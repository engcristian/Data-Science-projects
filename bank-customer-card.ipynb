># ACTIVE/CANCELED CARD DATA ANALYSIS CHART
 ># Plotting a bank card customer histogram from a dataset and collecting informations that reveals about activated and canceled customers. The dataframe is in Portuguese, but impotant to know that the blue charts are the active customers ('Cliente') and the red one are canceled ('Cancelado').

```python
import pandas as pd

bank_df = pd.read_csv('customers_bank.csv', encoding='latin1')
display(bank_df)
```

```python
bank_df = bank_df.drop('CLIENTNUM', axis = 1)
display(bank_df)
```

># Checking the dataframe info is important to know general informations

```python
display(bank_df.info())
```

># After checking it's necessary delete all blank elements

```python
bank_df = bank_df.dropna()
display(bank_df.info())
```

># Describing the values inside the dataframe

```python
display(bank_df.describe())
```

># Analyzing the numbers by any columns with value_counts() or % with value_counts(True).
Just inserting the "column's name" 

```python
display(bank_df['Idade'].value_counts())
```

># Analyzing all  canceled and active customers by the column 'Categoria'and 'Idade'.

```python
import plotly.express as px
```

```python
client_fig = px.histogram(bank_df, x= 'Idade', color ='Categoria')
client_fig.show()
```

># To analyse all columns is necessary use a repetition with FOR.

for column in bank_df:
    client_fig = px.histogram(bank_df, x = column, color = 'Categoria')
    client_fig.show()
